## MATPLOTLIB

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[5,15,35]
y=[100,250,725]
```

```python
plt.plot(x,y)
plt.show()
```

```python
fruit_prices=pd.DataFrame({"fruits":["Apple","Banana","Orange","Lemon","Pineapple"],
                          "prices":[5,11,6,3,12]})
fruit_prices
```

```python
plt.scatter(fruit_prices["fruits"],fruit_prices["prices"])
```

```python
plt.plot(fruit_prices["fruits"],fruit_prices["prices"],color="#AB4FE7",marker="*")
plt.title("FRUITS PRICES")
plt.xlabel("fruits")
plt.ylabel("prices")
```

## DATA VISUALIZATION

```python
df=pd.read_csv("leafshape.csv")
import seaborn as sns
```

```python
plt.figure(figsize=(12,7))
sns.distplot(df["bladelen"],color="#AB4FE7") #if you want,you can add this -- kde = "false" --
```

```python
plt.figure(figsize=(12,7))
sns.countplot(x="location",data=df)
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x="location",y="loglen",data=df)
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x="location",y="bladelen",data=df,hue="location",palette="Dark2")
```
